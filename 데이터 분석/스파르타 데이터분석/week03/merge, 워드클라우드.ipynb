```python
import pandas as pd
```

```python
enroll = pd.read_csv('./data/enrolleds_detail.csv')
enroll
```

```python
enroll_detail = enroll.groupby('lecture_id')['user_id'].count()
```

```python
enroll_detail
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = "Malgun Gothic"

plt.figure(figsize = (22,5))
plt.bar(enroll_detail.index, enroll_detail)
plt.title('강의에 따른 수강완료 수의 합계')
plt.xticks(rotation = 90)
plt.show()
```

```python
lectures = pd.read_csv('./data/lectures.csv')
lectures.tail(5)
```

```python
lecture_count =pd.DataFrame(enroll_detail)
lecture_count
```

```python
lecture_count=lecture_count.reset_index()
lecture_count
```

```python
lecture_count=lecture_count.rename(columns = {'user_id':'count'})
lecture_count
```

```python
lectures = lectures.set_index('lecture_id') #index로 설정
lectures
```

```python
full_lecture =lecture_count.join(lectures, on = 'lecture_id')
```

```python
full_lecture
```

```python
plt.figure(figsize = (22,10))
plt.bar(full_lecture['title'], full_lecture['count'])
plt.title('강의에 따른 수강완료 수의 합계')
plt.xlabel('강의명')
plt.xticks(rotation = 90)
plt.show()
```

```python
#워드 클라우드 
import numpy as np
from PIL import Image
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = open('./data/Sequence_01.txt', encoding='UTF8')
text = text.read()
text = text.replace('\n'," ")
text
```

```python
def file_open(number):
    text = open('./data/Sequence_%02d.txt' % number, encoding='utf-8-sig')
    text = text.read()
    text = text.replace('\n'," ")
    return text

result = ""
for i in range(1,14):
    result += file_open(i)
    
```

```python
# 특수 기호제거 
import re
pattern = '[^\w\s]'
result = re.sub(pattern= pattern, repl='', string = result)
result
```

```python
import matplotlib.font_manager as fm

for f in fm.fontManager.ttflist:
    if 'Gothic' in f.name:
        print(f.fname)
```

```python
font_path = 'C:\Windows\Fonts\H2GTRM.TTF'
wc = WordCloud(font_path = font_path, background_color='white')
wc.generate(result)
plt.figure(figsize=(50,50))
plt.axis('off') #축이 없다 
plt.imshow(wc)
plt.show()
```

```python
#Generate a word cloud image
mask = np.array(Image.open('./data/sparta.png'))

wc = WordCloud(font_path = font_path, background_color='white', mask= mask)
wc.generate(result)

f = plt.figure(figsize=(50,50))
f.add_subplot(1,2,1)
plt.imshow(mask, cmap = plt.cm.gray)
plt.title('Original Stencil', size = 40)
plt.axis('off')
f.add_subplot(1,2,2)

plt.imshow(wc, interpolation='bilinear')
plt.title('Sparta Cloud', size = 40)
plt.axis('off')
plt.show()
```
