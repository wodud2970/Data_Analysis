# 유동인구 + 상권데이터 합치기

```python
chicken_count_gu = seoul_chicken_data.groupby('구')['상권업종소분류명'].count().sort_values(ascending=False)
data = pd.DataFrame(chicken_count_gu).reset_index()
data
```

```python
sum_of_population_by_gu=population.groupby('군구')['유동인구수'].sum()
data1=pd.DataFrame(sum_of_population_by_gu).reset_index()
data1
```

# Merge, Join, Concat


```python
gu_chicken=data.join(data1.set_index('군구'), on = '구')
gu_chicken
```

```python
gu_chicken['유동인구수/치킨집수']=gu_chicken['유동인구수'] / gu_chicken['상권업종소분류명']
gu_chicken = gu_chicken.sort_values(by='유동인구수/치킨집수')
gu_chicken
```

```python
#유동인구수가 많고 치킨집수가 작을수록 값이 커지기 때문에 송파구가 가장 적합하다 
plt.figure(figsize=(10,5))
plt.bar(gu_chicken['구'],gu_chicken['유동인구수/치킨집수'])
plt.title('치킨집당 유동인구수')
plt.xlabel('구')
plt.ylabel('유동인구수/치킨집수')
plt.xticks(rotation = -45)
plt.show()
```

```python
#4월과 7월의 강남구의 일별 유동인구
four_data = pd.read_csv('./data/population04.csv')
seven_data = pd.read_csv('./data/population07.csv')
all_data = four_data.append(seven_data)
```

```python
four_data = pd.read_csv('./data/population04.csv')
seven_data = pd.read_csv('./data/population07.csv')

four_data = four_data[four_data['군구'] == '강남구'].groupby('일자').sum()
seven_data = seven_data[seven_data['군구'] == '강남구'].groupby('일자').sum()


plt.figure(figsize= (20,5))
plt.plot(four_data.index.astype('str'), four_data['유동인구수'])
plt.plot(seven_data.index.astype('str'), seven_data['유동인구수'])

plt.xlabel('날짜')
plt.title('2020년 4월 + 7월 서울 강남국 날짜별 유동인구수')
plt.xticks(rotation = -90)
plt.show()
```
