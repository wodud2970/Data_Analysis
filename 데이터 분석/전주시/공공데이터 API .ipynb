```python
# !pip install xmltodict
import requests, xmltodict
import json
import pandas as pd
import numpy as np

url = 'http://openapi.jeonju.go.kr/rest/childrenprotectionzone/getChildrenProtectionZone'
encoding = "GqzSDZ5bc1k%2F3vm47DJm29ZWdF%2FaUn5ya2HZV2vhVPXSyRS13FKhMBNH8bObRhEC8raIy68z3D0Q2JKbglk3KQ%3D%3D"
decoding = "GqzSDZ5bc1k/3vm47DJm29ZWdF/aUn5ya2HZV2vhVPXSyRS13FKhMBNH8bObRhEC8raIy68z3D0Q2JKbglk3KQ=="
result = url + "?serviceKey=" + encoding + "&authApiKey=" + decoding + "&startPage="

lst = []
page = 0

while True:
    try:
        content = requests.get(result+ "%s" %(page)).content
        dict = xmltodict.parse(content)
        jsonString = json.dumps(dict['rfcOpenApi']['body']['data']['list'], ensure_ascii=False)
        a = json.loads(jsonString)
        lst.append(a)
        page += 1
        print(page)
    except:
        break
    
    

all_obj = []
for i in range(len(lst)):
    for j in range(len(lst[i])):
        all_obj.append(lst[i][j])
        
#데이터 프레임 저장
title = "방구"
df = pd.DataFrame(data = all_obj)
df.to_csv(title+ ".csv",encoding="utf-8")
df
```
