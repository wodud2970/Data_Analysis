### 피쳐의 정규화
#### 연속형 피쳐 정규화
- 피쳐 스케일링 (키와 몸무개로 나이를 예측) 키와 몸무개의 단위가 다르다 따라서 피쳐들의 단위를 0~1사이 혹은 상대적 값을 표현하는 수치로 맞추어 주어야한다 

```python
#투수의 연봉에측하기
#scientific-notation이 아닌 float 모양으로 출력
pd.options.mode.chained_assignment = None

#피여 각각에 대한 스케일링을 수행하는 함수 
def standard_scaling(df, scale_columns):
    for col in scale_columns:
        series_mean = df[col].mean()
        series_std = df[col].std()
        df[col] = df[col].apply(lambda x : (x-series_mean)/series_std)
    return df

scale_columns = list(picher.columns[2:-2])+ list(picher.columns[-1:])
picher_df = standard_scaling(picher, scale_columns)
#피처 스케일링
```

```python
picher_df = picher_df.rename(columns={'연봉(2018)':'y'})
picher_df.head()
```

#### 범주형 피쳐 정규화
- 원-핫 인코딩 적용 get_dummies()

```python
#팀명 피처를 원핫 인코딩으로 변환
team_encoding =pd.get_dummies(picher_df['팀명'])
#팀명 열날리고
picher_df = picher_df.drop('팀명', axis = 1)
#join 을 사용하여 합쳐준다 
picher_df = picher_df.join(team_encoding)
team_encoding
#범주형 피처의 변환
```

```python
#train_test_split
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
#학습 데이터와 테스트 데이터로 분리 
print(type(picher_df.columns))
#선수명과 y를 날린다 
X = picher_df[picher_df.columns.difference(['선수명', 'y'])]
y = picher_df['y']
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size = 0.2,random_state = 19)
#데이터의 분리
```

```python
#회귀분석 계수로 학습
lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train)

print(lr.coef_)
```

```python
#모델 평가하기 (유의미한값)
import statsmodels.api as sm
X_train =sm.add_constant(X_train)
model = sm.OLS(y_train,X_train).fit()
model.summary()
# 결정계수 (R-squared) or 수정 결정계수(Adj, R-squared) 점수를 봐야한다  : 회귀분석으로 추정한 모델이 주어진 데이터를 얼마나 잘 설명하는가에 대한 점수 
# F 통계량(F-statistic) : 회귀식의 유의성 검정에 사용되는 값으로 F통계량에 대한 p-value인 Prob(F-static) 수치와 함께 살펴봐야한다 
# p-value가 0.05dlgkaus : F 통계량이 유의한 의미를 가진다 결론을 내려도 무방하다 
# p>|t|라는 정보는 각 피쳐 검정통계량(t-statics)이 얼마나 유의미한지에 대한 p-value를 나타낸 것이다 
#즉 'WAR', "연봉(2017)", "한화" 3개의 피쳐의 p-value가 0.05미만으로 나타나기 때문에 회귀분석에 유의미한 피처들이라는 결론이다 
```

```python
#어떤 피쳐가 가장 영향력이 강한 피처일까 
#한글 출력을 위한 사전 설정 
import matplotlib as mpl
mpl.rc('font', family = 'gulim')
plt.rcParams['figure.figsize'] = [20,16]
```

```python
#회귀 계수를 리스트로 반환 (영향력이 높은값)
coefs = model.params.tolist()
coefs_series = pd.Series(coefs)

#변수명을 리스트로 반환
x_labels =model.params.index.tolist()

#회귀계수를 출력
ax= coefs_series.plot(kind = 'bar')
ax.set_xlabel('x_features')
ax.set_ylabel('coef')
ax.set_xticklabels(x_labels)

#즉 'WAR', "연봉(2017)", "한화" 3개의 피쳐의 p-value가 0.05미만으로 나타나기 때문에 회귀분석에 유의미한 피처들이라는 결론이다 
# FIP : 유의미한 피쳐 :x  영향력이 큰 피쳐:o
# WAR : 유의미한 피쳐 :o  영향력이 큰 피쳐:o
# 홈런 : 유의미한 피쳐 :x  영향력이 큰 피쳐:o
# 작년연봉 : 유의미한 피쳐 :o  영향력이 큰 피쳐:o
# 한화 : 유의미한 피쳐 :o  영향력이 큰 피쳐:세모
```

```python
#R2 score
#학습 데이터와 테스트 데이터로 분리 
X = picher_df[picher_df.columns.difference(['선수명','y'])]
y = picher_df['y']
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size = 0.2,random_state = 19)

lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train)

#회귀 분석 모델 평가
#R2 score
print(model.score(X_train, y_train)) 
print(model.score(X_test, y_test)) 
#R2score
```

하지만 회귀분석은 독립성을 전제로한다  따라서 올바른 회귀 분석을 하려면 피처 쌍을 제거해야한다 
>다중공선성을 살펴봐야한다 

>다중공선성 : 변수 간 상관 관계가 높아 분석에 부정적인 영향을 미치는 것 (피처를 제거를 옳은 것인지 아닌지 판단의 기준을 제시
분산팽창요인(VIF)이라는 계수로 평가 VIF 10~15정도를 넘으면 피처는 다중 공선성 문제가 발생

다중공선성 코드 = [variance_inflation_factor(X.values,i) for i in range(X.shape[1])] 피처마다 variance_inflation_factor()함수를 적용

```python
#분산 팽창성을 통한( 상관 관계가 높아 분석에 부정적 영향을 미치는 것 제거)
from statsmodels.stats.outliers_influence import variance_inflation_factor 

#피쳐마다 VIF 계수를 출력 
vif = pd.DataFrame()
vif['VIF Factor'] = [variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
vif['features'] = X.columns
vif.round(1)
#WAR, 연봉(2017),FIP, WAR,볼넷/9, 삼진/9,홈런/9 : 회귀식에 큰 영향을 미치고 
#WAR, 연봉 2017: 유의미한 피쳐 
#FIP, kFIP : 상호 연관성을 가진 피처들은 높은 VIF수치를 보인다는 것을 알 수 있다 
#사용할 피처들을 선별
#다중 공산성
```
