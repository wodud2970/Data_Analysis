### NetworkX를 사용하여 단어 동시 출현 분석

```python
import pandas  as pd
from tqdm import tqdm

if __name__ == '__main__':
    #단어쌍의 빈도를 체크하기 위해 DTM을 불러온다
    dataset = pd.read_csv('Uljugun_news_content.csv')
    
    #단어들의 목록을 가져온다
    #이때 0번째 인덱스에는 빈 칸이 들어오므로 인덱싱을 없애준다
    column_list = dataset.columns[1:]
    #총 단어의 열값 갯수 
    word_length = len(column_list)
    
    #각 단어쌍의 빈도수를 저장할 dictionary 생성
    count_dict = {}
    #첫번째 단어기준으로 두번째 단어가 존재 할 경우 
    for doc_number in tqdm(range(len(dataset)), desc = '단어쌍 만들기 진행'):
        tmp = dataset.loc[doc_number] #한 문장 ( 현재 문서의 단어 출현 빈도 데이터를 가져온다 )
        for i, word1 in enumerate(column_list):
            if tmp[word1]: #한 문장에 단어가 존재하면 #현재문서에 첫번째 단어가 존재할 경우 
                for j in range(i+1, word_length): #문자안에 단어의 순서값
                    if tmp[column_list[j]]: #현재 문서에 두번째 단어가 존재할 경우 
                        count_dict[column_list[i], column_list[j]] = count_dict.get( (column_list[i], column_list[j]), 0) + max(tmp[word1], tmp[column_list[j]])
    # count_list 에 word1, word2, frequency형태로 저장 할 것이다
    count_list = []
    
    for words in count_dict:
        count_list.append([words[0], words[1], count_dict[words]])
    # 단어쌍 동시 출현 빈도를 DataFrame 형식으로 반든다
    df = pd.DataFrame(count_list, columns = ['word1', 'word2', 'freq'])
    df = df.sort_values(by = ['freq'], ascending = False)
    df = df.reset_index(drop = True)
    
    # 작 업이 오래 걸리기 때문에 csv파일로 저장후 사용
    df.to_csv('networkx.csv', encoding = 'utf-8-sig')
    
```

- DTM의 행은 문서의 번호, 열은 각 단어의 의미를 해당 문서에 각 단어가 출현한 빈도수를 나타내는것
- count_dict : 모든 문서에 대해 출현한 단어 쌍의 빈도수를 계산할 수 잇다

### 네트워크 분석

- 연관어, 관련어 한 쌍이 함께 등장하는 횟수 : 간선
- 해당하는 단어 하나는 : 노드
- 단어가 등장하는 횟수 : 노드 크기

- 연결중심성(Degree Centrality)
    - 연결된 노드가 많을 수록 중심성의 크기가 커진다는 관점
    - 연관어의 개수를 통해 해당 단어가 얼마나 중요한지 노드의 크기를 통해나타 낼 수 있다
- 매개중심성(Betweenness Centrality)
    - 노드와 노드 사이의 최단 경로를 계산할 때 해당 노드를 얼마나 거쳐가는 가를 척도로 한다
- 근점중심성(Closeness Centrality)
    - 다른 노드들까지의 최단 경로가 가까울수록 해당 노드가 중요하다는 척도를 가진다
    - 여기서는 연관어들 사이에 얼마나 연관성이 깊은지를 판단할 수 있을 것이다
- 고유 벡터 중심성(Eigenvector Centrality)
    - 연결 중심성과 약간 비슷
    - 차이가 있다면 연결된 다른 노드의 중심성도 고려한다는 것
    - 중심성의 크기가 큰 노드들과 많이 연결될수록 중요하다고 판단한다
- 페이지 랭크(Page Rank)
    - 다른 노드의 중심성과 연결되어있는 다른 노드의 개수에 따라 해당 노드의 중심성 크기를 상대적으로 결정하는 알고리즘
    - 여기서 상대적으로 결정한다는 것은 한 노드의 중심성이 A라고 한다면 연결되어있는 다른 노드의 중심성의 크기를 결정할 때 일정한 임의의 수로 나눠 A / x의 중심성을 가지게 된다
    - 지금까지 등장한 알고리즘중 가장 최적화된 알고리즘

```python
import pandas as pd
import networkx as nx
import operator
import numpy as np
if __name__ == '__main__':
    #단어쌍 동시출현 빈도수를 담았던 networkx.csv를 불러온다 
    data = pd.read_csv('networkx.csv')
    
    #중심성 척도 계산을 위한 Graph를 만든다 
    G_centrality = nx.Graph()
    
    #빈도수가 20000 이상인 단어에 대해서만 edge(간선)을 표현 ..(난 일단 작아서 300으로 한다)
    for ind in range(len(np.where(data['freq'] >= 100)[0])):
        G_centrality.add_edge(data['word1'][ind], data['word2'][ind], weight = int(data['freq'][ind]))
    
    dgr = nx.degree_centrality(G_centrality) #연결 중심성
#     btw = nx.betweenness_centrality(G_centrality) #매개 중심성
    cls = nx.closeness_centrality(G_centrality) #근접 중심성
    egv = nx.eigenvector_centrality(G_centrality) #고유벡터 중심성
    pgr = nx.pagerank(G_centrality) #페이지 링크
    
    sorted_dgr = sorted(dgr.items(), key = operator.itemgetter(1), reverse = True)
#     sorted_btw = sorted(btw.items(), key = operator.itemgetter(1), reverse = True)
    sorted_cls = sorted(cls.items(), key = operator.itemgetter(1), reverse = True)
    sorted_egv = sorted(egv.items(), key = operator.itemgetter(1), reverse = True)
    sorted_pgr = sorted(pgr.items(), key = operator.itemgetter(1), reverse = True)
    # 단어 네트워크를 그려줄 Graph 선언
    G = nx.Graph()
    
    #페이지 랭크에 따라 두 노드 사이에 연관성을 결정 (단어쌍의 연관성)
    #연결 중심성으로 계산 척도에 따라 노드의 크기가 결정된다 (단어의 등장 빈도수)
    #노드
    for i in range(len(sorted_pgr)):
        G.add_node(sorted_pgr[i][0], nodesize = sorted_dgr[i][1])
    #간선 
    for ind in range(len(np.where(data['freq'] > 100)[0])):
        G.add_weighted_edges_from([(data['word1'][ind], data['word2'][ind], int(data['freq'][ind]))])
    
    #노드 크기 조정
    sizes = [G.nodes[node]['nodesize'] * 1000 for node in G]
    
    options = {
        'edge_color': '#FFDEA2',
        'width': 1,
        'with_labels': True,
        'font_weight': 'regular'
        
    }
    
    #폰트 설정을 위한 font_manager import 
    import matplotlib.font_manager as fm
    import matplotlib.pyplot as plt
    
    fm._rebuild() 
    font_fname = 'C:\\Windows\\Fonts\\a한글나라BM.otf'
    fontprop = fm.FontProperties(fname = font_fname, size = 18).get_name()
    
    nx.draw(G, node_size = sizes, pos = nx.kamada_kawai_layout(G, scale =50), **options, font_family = fontprop) 
    ax = plt.gca()
    ax.collections[0].set_edgecolor('#555555')
    plt.rcParams['figure.figsize'] =(20,20)
    plt.savefig('content_network.jpg')
#     plt.show()
        
```

```python
#함수형으로 만든다 
import pandas as pd
from konlpy.tag import Okt
from sklearn.feature_extraction.text import CountVectorizer
from tqdm import tqdm
import re
import pandas as pd
import networkx as nx
import operator
import numpy as np


def make_DTM(data, file_name):
    
    def preprocessing(x):
        txt = re.sub('[\!@#$%\^&\*\(\)\-=_\[\]\{\}\.,/\?~\+\'"|]',"", x)
        return txt

    def hangul(x):
        hangul = re.compile('[^ ㄱ-ㅣ가-힣 a-z A-Z 0-9]')
        result = hangul.sub('',x)
        return result
    
    tagger = Okt()
    cv = CountVectorizer()
    
    #각 문서들의 말뭉치 (corpus)를 저장할 리스트 선언
    corpus = []
    for doc_num in tqdm(range(len(data)), desc = '말뭉치 만드는둥'):
        #한자와 특수문자 제거 
        txt = preprocessing(str(data.loc[doc_num]))
        txt = hangul(txt)
        noun_list = tagger.nouns(txt)
        corpus.append(' '.join(noun_list))
    DTM_Array = cv.fit_transform(corpus).toarray()
    feature_names = cv.get_feature_names()
    DTM_DataFrame = pd.DataFrame(DTM_Array, columns = feature_names)
    DTM_DataFrame.to_csv('%s.csv' % file_name, encoding = 'utf-8-sig')
    
    #단어쌍 만들기 
    column_list = DTM_DataFrame.columns
    word_length = len(column_list)
    
    count_dict = {}
    for doc_number in tqdm(range(len(DTM_DataFrame)), desc = '단어쌍 만들기 진행'):
        tmp = DTM_DataFrame.loc[doc_number] #한 문장 ( 현재 문서의 단어 출현 빈도 데이터를 가져온다 )
        for i, word1 in enumerate(column_list):
            if tmp[word1]: #한 문장에 단어가 존재하면 #현재문서에 첫번째 단어가 존재할 경우 
                for j in range(i+1, word_length): #문자안에 단어의 순서값
                    if tmp[column_list[j]]: #현재 문서에 두번째 단어가 존재할 경우 
                        count_dict[column_list[i], column_list[j]] = count_dict.get( (column_list[i], column_list[j]), 0) + max(tmp[word1], tmp[column_list[j]])
    count_list = []
    for words in count_dict:
        count_list.append([words[0], words[1], count_dict[words]])
    #단어쌍 동시 출현빈도를 DataFrame형으로 만든다 
    df = pd.DataFrame(count_list, columns = ['word1', 'word2', 'freq'])
    df = df.sort_values(by = ['freq'], ascending = False)
    df = df.reset_index(drop = True)
    
    df.to_csv('%s_networkx.csv' % file_name, encoding = 'utf-8-sig')
    return df

def make_network(data, file_name, mode, scale):
    #단어쌍 동시 출현 빈도수를 담았던 networkx
    G_centrality = nx.Graph()
#     FG = nx.from_pandas_edgelist(data, source = 'word1', target = 'word2')
#     FG.nodes()
#     dgr = nx.degree_centrality(FG)
#     pgr = nx.pagerank(FG) #페이지 링크
    for ind in range(len(np.where(data['freq'] >= mode)[0])):
        G_centrality.add_edge(data['word1'][ind], data['word2'][ind])
    
    dgr = nx.degree_centrality(G_centrality) #연결 중심성
#   btw = nx.betweenness_centrality(G_centrality) #매개 중심성
    cls = nx.closeness_centrality(G_centrality) #근접 중심성
#     egv = nx.eigenvector_centrality(G_centrality) #고유벡터 중심성
    pgr = nx.pagerank(G_centrality) #페이지 링크

    sorted_dgr = sorted(dgr.items(), key = operator.itemgetter(1), reverse = True)
#   sorted_btw = sorted(btw.items(), key = operator.itemgetter(1), reverse = True)
    sorted_cls = sorted(cls.items(), key = operator.itemgetter(1), reverse = True)
#     sorted_egv = sorted(egv.items(), key = operator.itemgetter(1), reverse = True)
    sorted_pgr = sorted(pgr.items(), key = operator.itemgetter(1), reverse = True)
    
    # 단어 네트워크를 그려줄 Graph 선언
    G = nx.Graph()
    
    #노드 
    for i in range(len(sorted_pgr)):
        G.add_node(sorted_pgr[i][0], nodesize = sorted_dgr[i][1])
#         print(sorted_pgr[i][0],sorted_pgr[i][1] )
    #간선 
    for ind in range(len(np.where(data['freq'] > mode)[0])):
        G.add_weighted_edges_from([(data['word1'][ind], data['word2'][ind], int(data['freq'][ind]))])
        #노드 크기 조정
    sizes = [G.nodes[node]['nodesize'] * 1000 for node in G]
    
    options = {
        'edge_color': '#FFDEA2',
#         'width': 1,
        'with_labels': True,
        'font_weight': 'regular'
        
    }
    
    #폰트 설정을 위한 font_manager import 
    import matplotlib.font_manager as fm
    import matplotlib.pyplot as plt
    
    fm._rebuild() 
    font_fname = 'C:\\Windows\\Fonts\\a한글나라BM.otf'
    fontprop = fm.FontProperties(fname = font_fname, size = 18).get_name()
    
    nx.draw(G, 
            node_size = sizes,
            pos = nx.spring_layout(G,k=3.5, iterations = 100, scale = scale),
            **options, 
            font_family = fontprop,
            node_shape='h',
            cmap=plt.cm.Blues, 
            alpha=0.9
           ) 
    ax = plt.gca()
    ax.collections[0].set_edgecolor('#555555')
    plt.rcParams['figure.figsize'] =(10,10)
    plt.savefig('%s_network.jpg' % file_name)
    
    
    
```
